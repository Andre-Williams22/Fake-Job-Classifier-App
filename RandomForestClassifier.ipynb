```python
# Grab Data from Github URL
url = 'https://raw.githubusercontent.com/Andre-Williams22/Fake-Job-Classifier-App/master/Dataset/fake_job_postings.csv'
```

```python
df['department'].fillna('missing', inplace=True)
df['location'].fillna('missing', inplace=True)
df['description'].fillna('missing', inplace=True)
df['company_profile'].fillna('missing', inplace=True)
df['requirements'].fillna('missing', inplace=True)
df['benefits'].fillna('missing', inplace=True)
df['employment_type'].fillna('missing', inplace=True)
df['required_experience'].fillna('missing', inplace=True)
df['industry'].fillna('missing', inplace=True)
df['function'].fillna('missing', inplace=True)
df['required_education'].fillna('missing', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Remove rows with missing salary range values 
df.dropna(subset=['salary_range'], inplace=True)
```

```python
len(df)
```

### Convert Data into numbers

```python
# use pd.get dummies 
X = pd.get_dummies(df.drop('fraudulent', axis=1))
y = df['fraudulent']
```

## Split the data into training and test

```python
# Split the data into training and test 
from sklearn.model_selection import train_test_split

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# View shapes of data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# 2. Choose Model (Estimator)

```python
from sklearn.ensemble import RandomForestClassifier

# instantiate the model
rf_model = RandomForestClassifier(n_estimators=100)

# Fit the model 
rf_model.fit(X_train, y_train)
```

# 3. Make Predictions

```python
# Make Predictions 
y_preds = rf_model.predict(X_test) # predictions
y_preds[:10]
```

```python
np.array(y_test[:10]) # actual results
```

# 4. Evaluate the Random Forest Classifier (Use the patterns the model have learned)


```python
print('Random Forest Classifier',rf_model.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score

np.random.seed(42)

rf_model = RandomForestClassifier(n_estimators=100)
scoring = {'acc': 'accuracy',
           'prec_macro': 'precision_macro',
           'rec_macro': 'recall_macro',
           'f1_macro': 'f1_macro'}
scores = cross_validate(rf_model,X,y, cv=10, scoring=scoring)
# print(scores.keys())

print ('Accuracy score : %.3f' % scores['test_acc'].mean())
print ('Precisoin score : %.3f' % scores['test_prec_macro'].mean())
print ('Recall score : %.3f' % scores['test_rec_macro'].mean())
print ('F1 score : %.3f' % scores['test_f1_macro'].mean())
```

# 5. Experimentation => Hyperparameter Tuning

```python
np.random.seed(42)

for i in range(10, 100, 10):
    print(f"Try with {i} estimators...")
    rf_model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {rf_model.score(X_test,y_test)}")
    print("")
```
